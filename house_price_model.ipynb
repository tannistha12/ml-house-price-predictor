```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")

# Load dataset
data = pd.read_csv('train.csv')

# Peek at the data
data.head()

# Shape of dataset
print(f"Rows: {data.shape[0]}, Columns: {data.shape[1]}")

# Dataset info
data.info()
```

```python
# Visualize missing values
plt.figure(figsize=(10,6))
sns.heatmap(data.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Distribution of target variable (SalePrice)
plt.figure(figsize=(8,5))
sns.histplot(data['SalePrice'], kde=True, color='green')
plt.title('Distribution of House Sale Prices')
plt.xlabel('Sale Price')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation with SalePrice
corr = data.corr(numeric_only=True)['SalePrice'].sort_values(ascending=False)
print("Top 10 Features Correlated with SalePrice:\n")
print(corr[1:11])
```

```python
top_corr_features = corr[1:11].index
plt.figure(figsize=(10,8))
sns.heatmap(data[top_corr_features].corr(), annot=True, cmap='coolwarm')
plt.title("Top 10 Correlated Features Heatmap")
plt.show()
```
